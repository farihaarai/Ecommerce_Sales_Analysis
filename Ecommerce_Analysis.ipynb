```python
# Import necessary libraries
import mysql.connector
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
# Connect to MySQL
db = mysql.connector.connect(
    host="localhost",       # Replace with your host (e.g., "127.0.0.1")
    user="root",   # Your MySQL username
    password="1234", # Your MySQL password
    database="project"
)

# Check connection
if db.is_connected():
    print("Connected to MySQL!")
```

```python
# Create cursor object to execute sql queries
cur = db.cursor()
# Query the data from cybersecurity attacks
cur.execute("SELECT * FROM sales")
# After fetching data from the database we are storing it into Pandas DataFrame
df = pd.DataFrame(cur.fetchall(), columns=[desc[0] for desc in cur.description])
# Print the Data
print(df)
```

**1. What is the distribution of order statuses?**

```python
query = """
SELECT order_status, COUNT(*) AS count
FROM sales
GROUP BY order_status;
"""
cur.execute(query)
data = cur.fetchall()

df = pd.DataFrame(data, columns=["order_status", "count"])
plt.figure(figsize=(8, 4))
sns.barplot(x="order_status", y="count", data=df)
plt.xticks(rotation=90)
plt.xlabel("Order Status")
plt.ylabel("Count")
plt.title("Distribution of Order Statuses")
plt.show()
```

**2. Which cities have the highest number of customers?**

```python
query = """
SELECT customer_city, COUNT(*) AS count
FROM sales
GROUP BY customer_city
ORDER BY count DESC
LIMIT 10;
"""
cur.execute(query)
data = cur.fetchall()

df = pd.DataFrame(data, columns=["customer_city", "count"])
plt.figure(figsize=(10, 5))
sns.barplot(x="customer_city", y="count", data=df)
plt.xticks(rotation=45)
plt.xlabel("City")
plt.ylabel("Number of Customers")
plt.title("Top 10 Cities with Most Customers")
plt.show()
```

**3. What are the top 10 products by sales volume?**

```python
query = """
SELECT product_id, COUNT(*) AS sales
FROM sales
GROUP BY product_id
ORDER BY sales DESC
LIMIT 10;
"""
cur.execute(query)
data = cur.fetchall()

df = pd.DataFrame(data, columns=["product_id", "sales"])
plt.figure(figsize=(12, 6))
sns.barplot(x="product_id", y="sales", data=df)
plt.xticks(rotation=45)
plt.xlabel("Product ID")
plt.ylabel("Sales Volume")
plt.title("Top 10 Products by Sales Volume")
plt.show()
```

**4. What percentage of orders use each payment type?**

```python
query = """
SELECT payment_type, COUNT(*) AS count
FROM sales
GROUP BY payment_type;
"""
cur.execute(query)
data = cur.fetchall()

# Convert the result into a DataFrame
df = pd.DataFrame(data, columns=["payment_type", "count"])

# Plot a pie chart
plt.figure(figsize=(8, 8))
plt.pie(df["count"], labels=df["payment_type"], autopct="%.1f%%", startangle=140, colors=sns.color_palette("pastel"))
plt.title("Payment Type Distribution")
plt.show()
```

**5. How are freight costs distributed across orders?** 
fright cost means overall transportation cost

```python
query = """
SELECT freight_value
FROM sales;
"""
cur.execute(query)
data = cur.fetchall()

df = pd.DataFrame(data, columns=["freight_value"])
plt.figure(figsize=(10, 6))
sns.histplot(df["freight_value"], kde=True, bins=30)
plt.xlabel("Freight Value")
plt.ylabel("Frequency")
plt.title("Distribution of Freight Costs")
plt.show()
```

**6. Which product categories contribute the most to sales?**

```python
query = """
SELECT product_category, SUM(price) AS total_sales
FROM sales
GROUP BY product_category
ORDER BY total_sales DESC
LIMIT 10;
"""
cur.execute(query)
data = cur.fetchall()

df = pd.DataFrame(data, columns=["product_category", "total_sales"])
plt.figure(figsize=(12, 6))
sns.barplot(x="product_category", y="total_sales", data=df)
plt.xticks(rotation=90)
plt.xlabel("Product Category")
plt.ylabel("Total Sales")
plt.title("Top 10 Product Categories by Sales")
plt.show()
```

**9. What are the most frequent cities for sellers?**

```python
query = """
SELECT seller_city, COUNT(*) AS count
FROM sales
GROUP BY seller_city
ORDER BY count DESC
limit 10;
"""
cur.execute(query)
data = cur.fetchall()

df = pd.DataFrame(data, columns=["seller_city", "count"])
plt.figure(figsize=(12, 6))
sns.barplot(x="seller_city", y="count", data=df)
plt.xticks(rotation=45)
plt.xlabel("City")
plt.ylabel("Number of Sellers")
plt.title("Most Frequent Cities for Sellers")
plt.show()
```

**10. Which states have the highest freight charges?**

```python
query = """ 
SELECT customer_state, AVG(freight_value) AS avg_freight
FROM sales
GROUP BY customer_state
ORDER BY avg_freight DESC;
"""
cur.execute(query)
data = cur.fetchall()

df = pd.DataFrame(data, columns=["customer_state", "avg_freight"])
plt.figure(figsize=(10, 6))
sns.barplot(x="customer_state", y="avg_freight", data=df)
plt.xticks(rotation=45)
plt.xlabel("State")
plt.ylabel("Average Freight")
plt.title("States with Highest Freight Charges")
plt.show()
```

**11. What are the correlations between numerical variables in the dataset?**

```python
query = """
SELECT price, freight_value, payment_value, payment_installments, product_weight_g
FROM sales;
"""
cur.execute(query)
data = cur.fetchall()

# Convert the result into a DataFrame
df = pd.DataFrame(data, columns=["price", "freight_value", "payment_value", "payment_installments", "product_weight_g"])

# Compute the correlation matrix
correlation_matrix = df.corr()

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix of Numerical Variables")
plt.show()
```

**13. Which sellers have the highest total sales?**

**14. How many items are typically in an order?** (Customer purchase Frequency)

```python
query = """
SELECT order_id, COUNT(order_item_id) AS num_items
FROM sales
GROUP BY order_id;
"""
cur.execute(query)
data = cur.fetchall()

df = pd.DataFrame(data, columns=["order_id", "num_items"])
plt.figure(figsize=(8, 6))
sns.histplot(df["num_items"], bins=20, kde=True)
plt.xlabel("Number of Items")
plt.ylabel("Number of Orders")
plt.title("Number of Items per Order")
plt.show()
```

**17. What is the share of top 5 product categories by the number of orders?**

```python
query = """
SELECT product_category, COUNT(*) AS count
FROM sales
GROUP BY product_category
ORDER BY count DESC
LIMIT 5;
"""
cur.execute(query)
data = cur.fetchall()

# Convert the result into a DataFrame
df = pd.DataFrame(data, columns=["product_category", "count"])

# Plot a pie chart
plt.figure(figsize=(8, 8))
plt.pie(df["count"], labels=df["product_category"], autopct="%.1f%%", startangle=140, colors=sns.color_palette("bright"))
plt.title("Top 5 Product Categories by Order Count")
plt.show()
```

**18. Which states have the most orders placed?**

```python
query = """
SELECT customer_state, COUNT(*) AS count
FROM sales
GROUP BY customer_state
ORDER BY count DESC
LIMIT 10;
"""
cur.execute(query)
data = cur.fetchall()

# Convert the result into a DataFrame
df = pd.DataFrame(data, columns=["customer_state", "count"])

# Create a donut chart
plt.figure(figsize=(8, 8))
plt.pie(df["count"], labels=df["customer_state"], autopct="%.1f%%", startangle=140, colors=sns.color_palette("cool"))
centre_circle = plt.Circle((0, 0), 0.70, fc="white")
plt.gca().add_artist(centre_circle)
plt.title("Top 10 States by Order Count")
plt.show()
```
